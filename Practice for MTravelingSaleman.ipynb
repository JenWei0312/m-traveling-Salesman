```python
# This block is to genrate the total cost matrix

# It is the sum of two sources of costs: traveling cost (none-diagnal) and work cost (diagnal)
# Any none-feasible connection between two nodes is indicated by a large distance: OnewayD

import numpy as np
import random

# Number of nodes
n=8                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   
# Number of salesmam
m=2
# Size of the distance matrix for 1-salesmen. 
# We add a row/col to represent the TRAVELING distance between the deport and any node.
N=n+1
# TRAVELING Distance of any node to itself
SelfD=0
# A large number to represent infeasible connection, such as one-way street
InfeasibleD=2000

# Initialize a random square matrix
# This matrix will go through the following steps to become the TRAVELING distanec matrix 
# for our n-node, m-salesmen problem
np.random.seed(2)
b=np.random.randint(0,30,size=(N,N))

# Step 1 -- Make it symmetrical
distance = (b+b.T)/2

# Step2 -- randomly asiign a few one-way street/infeasible connection 
# by making the distance between two random nodes much bigger than the distance for feasible connections
for i in range(3):
    np.random.seed(i)
    c=random.sample(range(1,len(distance)), 2)
    distance[c[0],c[1]]=InfeasibleD
    
# Step3 -- No traveling distance from any node to itself
np.fill_diagonal(distance,SelfD)
print("The initial distance matrix is")
print(distance)
print("The shape of the initial distance matrix is")
print(distance.shape)
print("The size of the initial distance matrix is")
print(distance.size)
print("The length of the initial distance matrix is")
print(len(distance))
print(sep='\n')
print(sep='\n')

######################################################################
# From this point on, we turn it into a function
# Step4 -- Extend the distance matrix for 1-salesmen to m-salesmen
# Repeat the first rows of the distance m-1 times
for i in range(m-1):
    distance = np.append(distance,distance[0:1,:], axis=0)
# Repeat the first colume of the distance m-1 times
for j in range(m-1):
    distance = np.append(distance,distance[:,0:1], axis=1)
    
# Step 5 -- Modify the extended distance matrix, 
# so that traveling among the last m-1 nodes is prohibited
distance[:1,-(m-1):]=InfeasibleD
distance[-(m-1):,:1]=InfeasibleD
distance[-(m-1):,-(m-1):]=InfeasibleD
    
#  Print the final TRAVELING distance matrix    
TravelDis=distance
print("The extended distance matrix is")
print(TravelDis)
print("The shape of the extended distance is")
print(TravelDis.shape)
print(sep='\n')
print(sep='\n')



###########################################################
# We may want to cut it off below

# Do we want to add the cost/time requirement to the disagnal of the TRAVELING distance mztrx?
# Now we have the total cost matrx
np.random.seed(2)
WorkCost=np.random.randint(0,25,size=(1,N))
print("The cost for completing the task at each node is")
print(WorkCost)
print("The shape of the work cost is")
print(WorkCost.shape)
print(sep='\n')
print(sep='\n')

# Now, adding the work cost  array to the diagnal of the distance matrix (traveling cost)
# To get the total cost matrix
np.fill_diagonal(distance[:N,:N],WorkCost)
TotalCost=distance
print("The total cost is")
print(TotalCost)
print("The shape of the total cost matrix is")
print(TotalCost.shape)

# We may want to cut it off above
###########################################################

```
